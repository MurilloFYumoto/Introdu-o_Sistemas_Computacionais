# PORTA XOR 

A porta XOR que nos foi proposto é uma feita com três entradas. A sua função é retonar o valor 1 se a soma das entradas for ímpar ou retornar zero quando a somatória for par. Optei por fazê-la com transistores N-MOS.

circuit-20220719-1435.png

# Componentes



A porta XOR com duas entradas é normalmente composta por duas AND e uma OR, como mostra abaixo.

Porta XOR normal.png

Entretanto, uma porta XOR com três entradas fica um pouco diferente. Sendo composta por oito AND e três OR.
